```python
#Creating Empty List
Laptop_name = []
Price = []
OS = []
Display = []
Processor = []
Hard_Disk = []
RAM = []
Weight = []
Dimension = []
GPU = []
```

```python
#Extracting the tags having computer Names:
names = driver.find_elements(By.XPATH,"//span[@class='datahreflink']/h3")
for i in names:
    if i.text is None:
        Laptop_name.append(" ") 
    else:
        Laptop_name.append(i.text.replace('\n','').strip())
        
# Price:
price = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']/table/tbody/tr[8]/td[3]")
for i in price:
    if i.text is None:
        Price.append(" ") 
    else:
        Price.append(i.text.replace('\n','').strip())
        
#operating System:
os = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']/table/tbody/tr[4]/td[3]")
for i in os:
    if i.text is None:
        OS.append(" ") 
    else:
        OS.append(i.text.replace('\n','').strip())
    
#display
displays = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']/table/tbody/tr[3]/td[3]")
for i in displays:
    if i.text is None:
        Display.append(" ") 
    else:
        Display.append(i.text.replace('\n','').strip())
    
#processor
processors = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']/table/tbody/tr[2]/td[3]")
for i in processors:
    if i.text is None:
        Processor.append(" ") 
    else:
        Processor.append(i.text.replace('\n','').strip())
        
# Memory(Hard_disk and RAM)
memories = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']/table/tbody/tr[5]/td[3]")
for i in memories:
    if i.text is None:
        Hard_Disk.append("-")
        RAM.append("-")
    else:
        disk=i.text
        RAM.append(disk.split('&')[0])
        Hard_Disk.append(disk.split('&')[1])       

# Weight & Dimension       
weights = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']/table/tbody/tr[7]/td[3]")
for i in weights:
    if i.text is None:
        Dimension.append("-")
        Weight.append("-")
    else:
        disk=i.text
        Dimension.append(disk.split('&')[0])
        Weight.append(disk.split('&')[1])          
        
        
#processor
Graphics = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']/table/tbody/tr[6]/td[3]")
for i in Graphics:
    if i.text is None:
        GPU.append(" ") 
    else:
        GPU.append(i.text.replace('\n','').strip())
        

df=pd.DataFrame({"Laptop name":Laptop_name,
                "Price":Price,
                "OS":OS,
                "Display":Display,
                "Hard Disk":Hard_Disk,
                 "RAM":RAM,
                "Processor":Processor,
                "Weight":Weight,
                "Dimension":Dimension,
                "Graphical processor":GPU})
   
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the homepage of hostelworld.com
url = "https://www.hostelworld.com/"
driver.get(url)
```

```python
#locating the location search bar
search_loc = driver.find_element(By.XPATH,"//input[@class='native-input body-1-reg']")
# write Lonodn in search bar
search_loc.send_keys("London")
time.sleep(2)
```

```python
#select london
london = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button')
london.click()
time.sleep(2)
```

```python
# Make empty lists
Hostel_Name = []
Distance = []
overall_review = []
total_reviews = []
facilities = []
price = []
Rating = []
property_description = []
```

```python
# Make empty lists
Hostel_Name = []
Distance = []
overall_review = []
total_reviews = []
facilities = []
price = []
Rating = []
property_description = []
```

```python
while(True):
    
    # Hostel name
    names = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
    for name in names:
        Hostel_Name.append(name.text)
    time.sleep(2)
        
    # Distance from city centre
    dis = driver.find_elements(By.XPATH,'//span[@class="description"]')
    for d in dis:
        Distance.append(d.text)
    time.sleep(2)
        
    # Overall Review    
    review = driver.find_elements(By.XPATH,'//div[@class="keyword"]//span')
    for r in review:
        overall_review.append(r.text)
    time.sleep(2)
    
    # Total No of reviews     
    t_review = driver.find_elements(By.XPATH,'//div[@class="reviews"]')
    for tr in t_review:
        total_reviews.append(tr.text)
    time.sleep(2)
    
    # facilities
    service = driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
    for s in service:
        
        facilities.append(s.text)
    time.sleep(2)
    
    # Prices    
    prices = driver.find_elements(By.XPATH,'//div[@class="price-col"]')
    for p in prices:
        price.append(p.text)
    time.sleep(2)    
        
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
        
time.sleep(4)        
        
# Separate  Privates_From price  and  Dorms_From price
Privates_From = []
for i in range(0,len(price),2):
    Privates_From.append(price[i])
time.sleep(2)

Dorms_From = []
for i in range(1,len(price),2):
    Dorms_From.append(price[i])           
```

```python
# scrape Hostels URL
hostel_url = []

while(True):
    urls = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]//a')
    for url in urls:
        hostel_url.append(url.get_attribute("href"))
    time.sleep(2)    
        
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break


        
Rate = []
for page in hostel_url:
    driver.get(page)
    
    # Rating
    try:
        ratings = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[1]/section/div[6]/div/div[1]/div[1]/div[1]')
        Rate.append(ratings.text)
    except NoSuchElementException:
        Rate.append("No Rating")  
    time.sleep(2)
    
    
    # Property Description
    try:
        pd = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[1]/section/div[6]/div/div[2]/div[2]/div/div[2]')
        property_description.append(pd.text)
    except NoSuchElementException:
        property_description.append("No Description")  

    
time.sleep(2)        
# remove extra data from Rating     
all_text = []
for i in Rate:
    all_text.append(i.split())
time.sleep(2)

for i in all_text:
    Rating.append(i[0]) 
```

```python
#creating dataframe
df = pd.DataFrame({'Hostel Name':Hostel_Name,
                'Distance from city centre':Distance,
                'Overall Review':overall_review[3:],
                'Total Reviews':total_reviews,
                'Facilities':facilities,
                'Privates From Price':Privates_From,
                'Dorms From Price':Dorms_From,
                'Rating':Rating,
                'Property Description':property_description})
```
